# Step 5. Creation of train and test data sets using optimum parameters

```python
#splitting dataset into x and y.

y = df['AveragePrice']
x = df.drop(columns=['AveragePrice'])
```

```python
x.head() #showing all columns except AveragePrice column.
```

```python
y.head()  #Showing only AveragePrice column.
```

```python
# Data Scaling
#StandardScaler = (X-mean)/std
from sklearn.preprocessing import StandardScaler   

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled 
```

```python
# Split data into train and test. Model will be built on training data and tested on test data
from sklearn.model_selection import train_test_split  #importing requried lib.

x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.25,random_state=340)
y_train
```

# Step 6. Model training using ML algorithm

```python
# Model instatiating and training

from sklearn.linear_model import LogisticRegression #Importing required library.
from sklearn.metrics import classification_report
```

```python
# Model instantiating and training

from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
# To find Intercept
print(regression.intercept_)

# To find coefficients
print(regression.coef_)
```

```python
# Adjusted R2 Score 

regression.score(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```

```python
y_pred = regression.predict(x_test)
y_pred
```

```python
data = pd.DataFrame({'Acutal data':y_test,'Predicted data':y_pred})
data
```

# Prediction with Region parameter

```python
#spliting dataset into X and Y.


X = df.drop(columns=['region'])
Y = df['region']
```

```python
X.head()  #showing all columns except region column.
```

```python
Y.head()  #showing only region column.
```

```python
# Data Scaling
#StandardScaler = (X-mean)/std
from sklearn.preprocessing import StandardScaler   

scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
x_scaled 
```

```python
#Spliting data into train test split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=100)
y_train
```

# Model training using ML algorithm

```python
# Model instatiating and training

from sklearn.linear_model import LogisticRegression #Importing required lib.
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix

LR = LogisticRegression()
LR.fit(x_train,y_train)
predlr = LR.predict(x_test)

print(accuracy_score(y_test,predlr))

print(confusion_matrix(y_test,predlr))

print(classification_report(y_test,predlr))
```

```python
from sklearn.neighbors import KNeighborsClassifier   #Importing requried lib.
KNN = KNeighborsClassifier()
KNN.fit(x_train,y_train)
predknn = KNN.predict(x_test)

print(accuracy_score(y_test,predknn))

print(confusion_matrix(y_test,predknn))

print(classification_report(y_test,predknn))
```

```python
from sklearn.ensemble import RandomForestClassifier     # Importing required lib
RAN = RandomForestClassifier()
RAN.fit(x_train,y_train)
predran = RAN.predict(x_test)

print(accuracy_score(y_test,predran))

print(confusion_matrix(y_test,predran))

print(classification_report(y_test,predran))
```

```python
from sklearn.model_selection import cross_val_score  #importing requried library
```

```python
scr = cross_val_score(LR,X,Y,cv=3)
print("Cross validation score for Logistic regrssion:",scr.mean())
```

```python
scr = cross_val_score(RAN,X,Y,cv=3)
print("Cross validation score for RandomForestClassifier:",scr.mean())
```

```python
scr = cross_val_score(KNN,X,Y,cv=3)
print("Cross validation score for KNeighborClassifier:",scr.mean())
```
